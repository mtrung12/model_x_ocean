```python
from sentence_transformers import SentenceTransformer
import faiss, pandas as pd, numpy as np, os

MODEL_DIR   = "models/sbert_essays_finetuned"
OUTPUT_DIR  = "models/rag_artifacts"
TRAIN_CSV   = "data/split/essays/train.csv"

os.makedirs(OUTPUT_DIR, exist_ok=True)
model = SentenceTransformer(MODEL_DIR, trust_remote_code=True)
model.max_seq_length = 1024

train_df = pd.read_csv(TRAIN_CSV)
texts = ["search_document: " + t for t in train_df["text"].tolist()]
emb = model.encode(texts, normalize_embeddings=True,
                   convert_to_numpy=True, show_progress_bar=True)

index = faiss.IndexFlatIP(emb.shape[1])
index.add(emb.astype(np.float32))
faiss.write_index(index, f"{OUTPUT_DIR}/train_index.faiss")
train_df.to_csv(f"{OUTPUT_DIR}/train_metadata.csv", index=False)
print("Done.")
```
