```python
from huggingface_hub import login
from kaggle_secrets import UserSecretsClient

user_secrets = UserSecretsClient()
token = user_secrets.get_secret("llama-3-8B")
login(token=token)
```

```python
from pathlib import Path
import sys

import pandas as pd

project_root = Path.cwd().resolve()
if not (project_root / "baselines").exists():
    project_root = (project_root / ".." / "..").resolve()

if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from baselines.evaluate import evaluate
from baselines.predict import predict
```

```python
test_df = pd.read_csv('/kaggle/working/model_x_ocean/data/split/essays/mini_test.csv')
model_name = "llama-3-8b"
temperature = 0
res_dir = '/kaggle/working/model_x_ocean/result'
log_dir = '/kaggle/working/model_x_ocean/log'
max_new_tokens = 512
```

```python
run_id, run_time, prediction_csv = predict(
    text_df=test_df,
    model_name=model_name,
    log_dir=log_dir,
    prompt_mode="zeroshot",
    max_new_tokens=max_new_tokens,
    res_dir=res_dir,
    temperature=temperature
)
```

```python
evaluation = evaluate(
    prediction_csv=prediction_csv,
    model_name=model_name,
    res_dir=res_dir,
    run_time=run_time,
    prompt_mode="zeroshot",
    run_id=run_id
)
```
